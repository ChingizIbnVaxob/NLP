```python
import os
from os import listdir
from os.path import isfile, join
from matplotlib import pyplot as plt
import scipy.io.wavfile as wav
import numpy as np
from keras.preprocessing import image
from keras.utils import to_categorical
```

```python
def wav_to_spectrogram(audio_path, save_path, spectogram_dimensions=(64,64), noverlap=16, cmap='gray_r'):
  sample_rate, samples = wav.read(audio_path)
  fig = plt.figure()
  fig.set_size_inches((spectogram_dimensions[0]/fig.dpi, spectogram_dimensions[1]/fig.dpi))
  ax = plt.Axes(fig, [0., 0., 1., 1.])
  ax.set_axis_off()
  fig.add_axes(ax)
  ax.specgram(samples, cmap=cmap, noverlap=noverlap)
  ax.xaxis.set_major_locator(plt.NullLocator())
  ax.yaxis.set_major_locator(plt.NullLocator())
  fig.savefig(save_path, bbox_inches='tight', pad_inches=0)
  plt.close()
```

```python
def dir_to_spectrogram(audi_dir, spectogram_dir, spectogram_dimensions=(64,64), noverlap=16, cmap='gray_r'):
  file_names = [f for f in listdir(audi_dir) if isfile(join(audi_dir, f)) and '.wav' in f]
  for file_name in file_names:
    print(file_name)
    audio_path = join(audi_dir, file_name)
    spectogram_path = spectogram_dir + file_name.replace('.wav', '.png')
    wav_to_spectrogram(join(audi_dir, file_name), join(spectogram_dir, file_name.replace('.wav', '.png')), spectogram_dimensions, noverlap, cmap)
```

```python
audio_folder = 'audio_uz/'
spectrogram_folder = 'spectrograms/'
dir_to_spectrogram(audio_folder, spectrogram_folder)
```
